# Cleanup

Deletes all Kafka topics created by the other notebooks.

```python
from kafka.admin import KafkaAdminClient
from config.kafka_config import *

admin = KafkaAdminClient(client_id='admin', **kafka_connection)

topics_to_delete = [
    topic_name,
    topic_name + '_partitioned',
    topic_name + '_schema',
]

existing = admin.list_topics()
to_delete = [t for t in topics_to_delete if t in existing]

if to_delete:
    admin.delete_topics(topics=to_delete, timeout_ms=timeout_ms)
    print('Deleted:', to_delete)
else:
    print('No topics to delete.')

admin.close()
```
